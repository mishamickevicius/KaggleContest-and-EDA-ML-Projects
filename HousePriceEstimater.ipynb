```python
def fill_null(data, fill_value, column):
    data[column] = data[column].fillna(fill_value)
    return data
```

```python
def dummies(data, column):
    dummies_train = pd.get_dummies(data[column], dtype=int)

    data = pd.concat([data, dummies_train], axis=1)

    data.drop(column, axis=1, inplace=True)

    return data
```

```python
def save_submission(pred, filename, X_test=X_test):
    answers = {'Id': X_test.index, 'SalePrice': pred}
    answers_df = pd.DataFrame(answers)
    answers_df.to_csv(filename, index=False)
    return 1
```
